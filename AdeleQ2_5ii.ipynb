```python
#Gideon Adele Q2 num 5 (Using tanh)
import numpy as np
import matplotlib.pyplot as plt
import random

#Reading in the datasets
x_train=np.loadtxt("/X_train.csv")
x_test=np.loadtxt("/X_test.csv")
y_test = np.loadtxt("/Y_test.csv")  
y_train = np.loadtxt("/Y_train.csv")
y=y_train.reshape(100, 1)
x=x_train.reshape(100,2)


number_of_neurons=5
w1=np.random.rand(number_of_neurons,2)
b1=np.random.rand(number_of_neurons,1)
w2=np.random.rand(1,number_of_neurons)
b2=np.random.rand(1,1)
learning_rate=1

learning_rate_range=np.random.uniform(0.001,1,25)
#sigmoid for output layer
def sigmoid (inp):
    return 1/(1+np.exp(-inp))
#using tanh for hidden layer
def tanh (x):
    return ((2/(1+np.exp(-2*x)))-1)

mse_list2=[]
w1_list=[]
w2_list=[]
b1_list=[]
b2_list=[]
t11_list=[]

for i in range (25):
    mse_list1=[]
    t1_list=[]
    t2_list=[]

    for j in range (100):
        #forward feed
        z1=np.dot(w1,x[j].reshape(2,1))+b1
        a1=tanh(z1)
        z2=np.dot(w2,a1)+b2
        a2=sigmoid(z2)

        aa=np.round(a2,0)
        print(a2)

        if y[j]==aa or a2:
            t1_list.append(1)
        else:
            t2_list.append(1)

        #back-prop
        dz2=a2-y[j].reshape(1,1)
        dw2=np.dot(dz2,a1.T)
        db2=dz2

        dz1=np.dot(w2.T,a2-y[j].reshape(1,1))*(1-a1*a1)
        dw1=np.dot(dz1,x[j].reshape(2,1).T)
        db1=dz1
        #updated rules
        w1=w1-learning_rate*dw1
        b1=b1-learning_rate*np.mean(db1)
        w2=w2-learning_rate*dw2
        b2=b2-learning_rate*np.mean(db2)
        #MSE
        mse= ((y[j]-a2)**2)
        mse_list1.append(mse)

    w1_list.append(w1)
    w2_list.append(w2)
    b1_list.append(b1)
    b2_list.append(b2)

    a=sum(mse_list1)/100
    mse_list2.append(a)
    b=(len(t1_list))/100
    t11_list.append(b)


print("Average MSE on Train data: ",sum(mse_list2)/25)
print("Average Accuracy on Train data: ",sum(t11_list)/25)


mse_list4=[]
t33_list=[]

#loop for test set
for i in range (len (w1_list)):
    t3_list=[]
    t4_list=[]
    mse_list3=[]
    for j in range (50):
        z1=np.dot(w1_list[i],x_test[j].reshape(2,1))+b1_list[i]
        a1=tanh(z1)
        z2=np.dot(w2_list[i],a1)+b2_list[i]
        a2=sigmoid(z2)

        aa=np.round(a2,0)

        if y[j]==aa or a2:
            t3_list.append(1)
        else:
            t4_list.append(1)

        #MSE
        mse= ((y[j]-a2)**2)
        mse_list3.append(mse)


    #print(t1,t2)
    a=sum(mse_list3)/50
    mse_list4.append(a)
    b=(len(t3_list))/50
    t33_list.append(b)
print("Average MSE on Test data: ",sum(mse_list4)/25)
print("Average Accuracy on Test data: ",sum(t33_list)/25)

L1=[k for i in mse_list2 for j in i for k in j]
plt.plot(np.arange(25),L1,'r-')
L1=[k for i in mse_list4 for j in i for k in j]
plt.plot(np.arange(25),L1,'g--')
plt.xlabel("Number of Iterations")
plt.ylabel("Mean Squared Error")
plt.legend(('Train Data','Test Data'))
plt.show()

plt.plot(np.arange(25),t11_list,'r')
plt.plot(np.arange(25),t33_list,'g')
plt.xlabel("Num of Iterations")
plt.ylabel("Accuracy")
plt.legend(('Train Data','Test Data'))
plt.show()

```
